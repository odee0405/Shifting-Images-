```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml

import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.ndimage.interpolation as ndi_int
```

```python
# Load Data
mnist = mnist = fetch_openml('mnist_784', version=1)
```

```python

X, y = mnist["data"], mnist["target"]

type(X), type(y)
```

```python
# convert into pandas data frame and series
X = pd.DataFrame(X)
y = pd.Series(y)
```

```python
type(X), type(y)
```

```python
digit = X.iloc[2,:].values
digit_img = digit.reshape(28,28)

plt.imshow(digit_img, cmap = mpl.cm.binary, interpolation="nearest")
plt.axis("off")
plt.show()
```

```python
# Shift image 5 pixels up
shifted_digit_img_test = ndi_int.shift(digit_img, shift=(-5, 0))
```

```python
# display the original and shifted images
fig, axes = plt.subplots(1, 2, figsize=(14, 14))
axes[0].imshow(digit_img, cmap=mpl.cm.binary)
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(shifted_digit_img_test, cmap=mpl.cm.binary)
axes[1].set_title('Shifted Image')
axes[1].axis('off')
plt.show()
```

```python
# Funtion to move image
def shift_image(digit_img,x,y):
    shifted_image = ndi_int.shift(digit_img,[x,y])
    fig, axes = plt.subplots(1, 2, figsize=(14, 14))
    axes[0].imshow(digit_img, cmap=mpl.cm.binary)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(shifted_image, cmap=mpl.cm.binary)
    axes[1].set_title('Shifted Image')
    axes[1].axis('off')
    plt.show()
    
    
```

```python
shift_image(digit_img,-5,0)
```
